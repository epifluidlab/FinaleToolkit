```python
import os
import finaletoolkit
from tqdm import tqdm
import subprocess
import glob
import numpy as np
```

```python
def read_ctcf_create_intervals(path, output):
    with open(path, 'rt') as ctcf, open(output, 'w') as output:
        for line in tqdm(ctcf):
            intervals = []
            chrom, start, end, _, _, strand = line.strip().split('\t')
            pos = (int(start) + int(end)) //2
            int_start = pos-2000
            int_end = pos+2001
            output.write(f"{chrom[3:]}\t{int_start}\t{int_end}\n")
```

```python
read_ctcf_create_intervals('../data/ctcf.hg19.bed', '../outputs/3E/ctcf_intervals.bed')
```
