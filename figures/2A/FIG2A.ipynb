```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
frag_len = pd.read_csv('../outputs/2A/BH01_fraglengths.unfiltered.tsv', sep='\t')
frag_len.sort_values(by='min', inplace=True)
frag_len = frag_len[(frag_len['min'] <= 300) & (frag_len['min'] >= 50)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create KDE plot
kde = sns.kdeplot(data=frag_len, x='min', weights='count', color='#D97D72', linewidth=2, bw_adjust=0.1)

# Scale the y-axis density by multiplying it by 100
kde_lines = kde.get_lines()[0]  # Get the first line of the KDE plot
kde_lines.set_ydata(kde_lines.get_ydata() * 100)  # Scale density values by 100

# Adjust x-axis and y-axis limits
plt.xlim(50, 300)
plt.ylim(-0.1, 1.5)

# Customize ticks
plt.xticks(range(50, 301, 50), fontsize=12)
plt.yticks([i * 0.5 for i in range(6)], fontsize=12)

# Labels and styling
plt.xlabel('Fragment sizes (bp)', fontsize=12)
plt.ylabel('Density', fontsize=12)  # Reflect multiplication in label
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Adjust figure size and resolution
plt.gcf().set_size_inches(4, 4)
plt.gcf().set_dpi(300)
plt.tight_layout()

# Show the plot
plt.show()
```
